```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('E:/creditcard.csv')
```

# Number of missing values in each column 

```python
df.isnull().sum()
```

```python
#Distribution of legit and fraudulent transactions
df['Class'].value_counts()
```

This dataset is highly unbalanced. Label 0 represents legit transactions and label 1 represents fraudelent transaction.

# Separating the data for analysis

```python
legit = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measures of the data
legit['Amount'].describe()
```

```python
fraud['Amount'].describe()
```

```python
#Comparing the values for both transactions
df.groupby('Class').mean()
```

As the dataset is highly unbalanced, the number of legit transactions are high, for that reason building a sample dataset containing similar distribution of legit transaction and fraud transactions
Number of fraud transactions = 492

```python
#Genereatinig random samples from the dataset
legitSample = legit.sample(n=492)
```

```python
#Concatinating 2 dataframes
new_df = pd.concat([legitSample, fraud], axis=0)
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_df.drop(columns='Class',axis=1)
Y = new_df['Class']
```

# Splitting the data into trainig data and testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

# Model Training: Logistic Regression

```python
model = LogisticRegression(max_iter=20000)
model.fit(X_train, Y_train)
```

# Model Evaluation: Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
